```python
import numpy as np
from PIL import Image
import os
```

```python
# --- Configuration ---
NPY_FILE = r'D:\Mix\Year4_1\Capstone\Experiment Data\31_10_25\1tick_3.npy'  # Path to your .npy file
OUTPUT_DIR = r"D:\Mix\Year4_1\Capstone\Experiment Data\1tick_3"   # Folder to save the PNGs
```

```python
# 1. Create the output directory if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
# 2. Load the numpy array
try:
    image_stack = np.load(NPY_FILE)
except FileNotFoundError:
    print(f"Warning: File '{NPY_FILE}' not found.")
    print("Creating a dummy array for demonstration...")
    # Create a dummy stack of 10 random 64x64 RGB images
    image_stack = (np.random.rand(10, 64, 64, 3) * 255).astype(np.uint8)

print(f"Array loaded. Shape: {image_stack.shape}, Dtype: {image_stack.dtype}")
```

```python
# 3. Check array shape
# We assume the first dimension is the number of images.
if image_stack.ndim not in [3, 4]:
    print(f"Error: Array must be 3D (N, H, W) or 4D (N, H, W, C).")
    exit()
```

```python
num_images = image_stack.shape[0]
print(f"Found {num_images} images to save...")
```

```python
# 4. Loop, convert, and save
for i in range(num_images):
    # Get the i-th image slice
    single_image_array = image_stack[i]

#--- IMPORTANT: Handle Data Type ---
#PIL requires uint8 (0-255). If your data is float (0.0-1.0)
#or another type, you MUST convert it. See the next section.
#For this example, we'll assume it's uint8.
    if single_image_array.dtype != np.uint8:
        print(f"Warning: Image {i} is {single_image_array.dtype}, not uint8.")
        print("See 'Handling Different Data Types' section below.")
        # Attempt to convert, but this may fail if data isn't 0-255
        try:
            single_image_array = single_image_array.astype(np.uint8)
        except ValueError:
            print(f"Could not convert image {i}. Skipping.")
            continue
```

```python
# 5. Create PIL Image and save
try:
    img = Image.fromarray(single_image_array)

#Define the output filename (e.g., image_0000.png, image_0001.png)
    output_filename = os.path.join(OUTPUTDIR, f'image{i:04d}.png')

#Save the image
    img.save(output_filename)

except ValueError as e:
    print(f"Error saving image {i}: {e}")
    print("This often happens if the array shape is wrong (e.g., RGB data in a 2D array).")

print(f"\nSuccessfully saved {num_images} images to {OUTPUT_DIR}")
```
